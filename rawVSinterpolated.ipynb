```python
import numpy as np
import matplotlib.pyplot as plt

```

```python
#omega, q: raw and ip are from data_M31 file
# from data_conv_M31 import kms_vcirc, kpc_r_Claude,g_Msun,cm_pc,s_Myr
from data.data_M31 import nandeleted_data,dat_omega,kpc_r_om_q,cm_km,cm_kpc

plt.plot((kpc_r_om_q)*(785/780),dat_omega/(cm_km/cm_kpc),label='raw',marker='.')
plt.plot(nandeleted_data[0],nandeleted_data[5]/(cm_km/cm_kpc),label='ip',marker='.')
plt.legend()
plt.xlabel('radius (kpc)')
plt.ylabel('omega (kms_kpc)')
plt.title('omega')
plt.show()

for i in range(len(kpc_r_om_q)):
    print(kpc_r_om_q[i],dat_omega[i])
print('*********************************************************')
for i in range(len(nandeleted_data[0])):
    print(nandeleted_data[0][i],nandeleted_data[5][i])
```

```python
#sigma_HI
from data.data_conv_M31 import Msunpc2_SigmaHI, kpc_r_cl,g_Msun,cm_pc
from data.data_M31 import nandeleted_data

sigma_HI=nandeleted_data[2]/(g_Msun/(cm_pc**2))
plt.plot(kpc_r_cl,Msunpc2_SigmaHI,label='raw')
plt.plot(nandeleted_data[0],sigma_HI,label='ip')
plt.legend()
plt.xlabel('radius (kpc)')
plt.ylabel('sigma_HI (M_sun/pc2)')
plt.title('sigma_HI')
plt.show()

for i in range(len(kpc_r_cl)):
    print(kpc_r_cl[i],Msunpc2_SigmaHI[i])
print('*********************************************************')
for i in range(len(nandeleted_data[0])):
    print(nandeleted_data[0][i],sigma_HI[i])
```

```python
#sigma_tot
from data.data_conv_M31 import Msunpc2_SigmaTot, kpc_r_cl,g_Msun,cm_pc
from data.data_M31 import nandeleted_data

sigma_tot=nandeleted_data[1]/(g_Msun/(cm_pc**2))
plt.plot(kpc_r_cl,Msunpc2_SigmaTot,label='raw',marker='.')
plt.plot(nandeleted_data[0],sigma_tot,label='ip',marker='.')
plt.legend()
plt.yscale("log")
plt.xlabel('radius (kpc)')
plt.ylabel('sigma_tot (M_sun/pc2)')
plt.title('Sigma total')
plt.show()

for i in range(len(kpc_r_cl)):
    print(kpc_r_cl[i],Msunpc2_SigmaTot[i])
print('*********************************************************')
for i in range(len(nandeleted_data[0])):
    print(nandeleted_data[0][i],sigma_tot[i])
```

```python
#sigma_SFR
from data.data_conv_M31 import Msunpc2Gyr_Sigma_SFR, kpc_r_SFR_TB10,g_Msun,cm_pc,s_Myr
from data.data_M31 import nandeleted_data

sigma_sfr=nandeleted_data[-2]/(g_Msun/((s_Myr*10**(3))*(cm_pc**2)))

plt.plot(kpc_r_SFR_TB10,Msunpc2Gyr_Sigma_SFR,label='raw',marker='.')
plt.plot(nandeleted_data[0],sigma_sfr,label='ip',marker='.')
plt.legend()
# plt.yscale("log")
plt.xlabel('radius (kpc)')
plt.ylabel('sigma_sfr (M_sun/pc2_Gyr)')
plt.title('Sigma sfr')
plt.show()

for i in range(len(kpc_r_SFR_TB10)):
    print(kpc_r_SFR_TB10[i],Msunpc2Gyr_Sigma_SFR[i])
print('*********************************************************')
for i in range(len(nandeleted_data[0])):
    print(nandeleted_data[0][i],sigma_sfr[i])
```
